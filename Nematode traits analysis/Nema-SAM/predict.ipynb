```python
from ultralytics import YOLO
```

```python
model = YOLO("best.pt")
```

```python
test_path = r'test_image\48.png'
results = model.predict(source=test_path, name='predict_output', save=True, show_conf=False, show_labels=False, line_width=1, boxes=True, iou=0.3, save_crop=False)
```

```python
result = results[0]
boxes = result.boxes
print(boxes.conf)
if boxes.conf == None:
    print("none")
```
